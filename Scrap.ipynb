```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
idx = '959.jpg'
img = cv2.imread('/Users/jhassan/Downloads/ut-interaction_set2/seq18/'+idx,1)
# cv2.imshow('Image', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
cv2.rectangle(img, (330,210), (370,240),[0,0,255], 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax = plt.gca()
ax.set_xticks(np.arange(0, 750, 50));
ax.set_yticks(np.arange(0, 500, 50));
ax.grid(color='w', linestyle='-', linewidth=0.5)
plt.show()
print(f'Shape of the image is {img.shape}')
```

```python
img_gs = img[:,512:,:]
img_rgb = img[:,:512,:]
plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_gs, cv2.COLOR_BGR2RGB))
plt.show()
```
